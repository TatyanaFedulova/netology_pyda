## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
# загружаем

power = pd.read_csv('power.csv')
power.head()
```

```python
# создаем отфильтрованный датасет со странами Прибалтики

power_Baltic = power[(power['country'] == 'Lithuania') | (power['country'] == 'Latvia') | (power['country'] == 'Estonia') ]
```

```python
power_Baltic['country'].unique()
```

```python
# фильтруем датасет по категориям (4, 12, 21) и годам (2005 - 2010), убираем отрицательные значения quantity

power_Baltic = power_Baltic[((power_Baltic['category'] == 4) | (power_Baltic['category'] == 12) | (power_Baltic['category'] == 21)) & (power_Baltic['year'] > 2004) & (power_Baltic['year'] < 2011) & (power_Baltic['quantity'] > 0) ]
power_Baltic.head (5)
```

```python
# суммируем quantity по категориям

power_Baltic['quantity'].sum()
```
